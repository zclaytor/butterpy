```python
import numpy as np
from butterpy import regions as regions1
from butterpy.core import regions as regions2
```

```python
seed = 89
np.random.seed(seed)
r1 = regions1(activity_rate=1, min_ave_lat=10, max_ave_lat=50, cycle_length=3, cycle_overlap=1, tsim=3600, butterfly=True)
np.random.seed(seed)
r2 = regions2(activityrate=1, minlat=10, maxlat=50, cyclelength=3, cycleoverlap=1, ndays=3600, butterfly=True)
```

```python
len(r1), len(r2)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
fig, ax = plt.subplots(2, 1, figsize=(8,8))
ax[0].plot(r1["nday"], r1["lat"], "ko", alpha=0.3)
ax[1].plot(r2["nday"], 90-(r2["thpos"]+r2["thneg"])*90/np.pi,"ko", alpha=0.3)
```

```python
r1.query("lat > 0").lat.max()
```

```python
r2.to_pandas().eval("lat = 90 - (thpos + thneg)*90/3.1416").query("lat > 0").lat.max()
```
